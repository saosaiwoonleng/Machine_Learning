# Imbalance dataset

The first set will contain the first 100 rows which are of loan_status=0 and the last 200 rows containing each 100 rows of loan_status=1 and loan_status=2. I will create the second dataset with the next 100 rows of loan_status=0 with the last 200 rows of the dataset. And finally the last 100 rows of loan_status=0 with the last 200 rows of the dataset. This will create 3 sets of data with equal distribution of category

```python
#getting all the 3 different data
df_set1 = df_change1[(df_change1.index<=99) | (df_change1.index>=300)]
df_set2 = df_change1[((df_change1.index>=100) &(df_change1.index <= 199)) | (df_change1.index>=300)]
df_set3 = df_change1[((df_change1.index>=200) &(df_change1.index <= 299)) | (df_change1.index>=300)]
#combining all 3 sets to make a 900 samples dataset to ensure equal weightage given to all category
df_med = df_set1.append(df_set2)
df_final = df_med.append(df_set3)
df_final
```

KNN for new dataset. Best accuracy = 50.89% with k =1

```python
from sklearn.decomposition import PCA

x = df_final[["age","education","Gender","pressure","Principal","terms"]]
y = df_final["loan_status"]

k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])

print("start time is:\n{0}".format(datetime.now().time()))
for i in range(1,31,+2):
    clf = KNeighborsClassifier(n_neighbors = i)
    scoring = 'accuracy'
    score = cross_val_score(clf, principalDf, y, cv=k_fold, n_jobs=1, scoring=scoring,error_score=np.nan)
    print(i)
    print(round(np.mean(score)*100, 2))
print("end time is:\n{0}".format(datetime.now().time()))

for i in range(1,31,+2):
    sum = 0
    knn_model = KNeighborsClassifier(n_neighbors = i)
    for j in range(0,50):
        x_train, x_test, y_train, y_test = train_test_split(principalDf, y, random_state = j)
        knn_model.fit(x_train, y_train)
        y_predict = knn_model.predict(x_test)
        
        sum += accuracy_score(y_test, y_predict)
    print(i)
    print(sum/50)
cm= confusion_matrix(y_test,y_predict)
print(cm)
```
